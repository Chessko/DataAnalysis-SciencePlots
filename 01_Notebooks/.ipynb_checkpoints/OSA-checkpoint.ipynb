```python
#Imports

import datetime
import numpy as np
import scipy as sp
import scipy.fftpack
import scipy.integrate
import scipy.signal
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Initiating DataFrame

# File doesn't exist anymore --> Notebook still helpful for future files
df0 = pd.read_csv('../01_Data/cwlaser_OSA.CSV',names=['Wavelength / nm', 'Power / dBm']) 
df = df0[30:].astype(float)
```

```python
from scipy.interpolate import UnivariateSpline
from scipy.optimize import leastsq

x = df["Wavelength / nm"]
y = df["Power / dBm"]

spline = UnivariateSpline(x, y-(np.max(y)-3), s=0)
r1, r2 = spline.roots() # find the roots
threedB_bandwidth = abs(r2-r1)
```

```python
centerwavelength = df["Wavelength / nm"][df["Power / dBm"].idxmax()]
```

```python
# Plot
x = df["Wavelength / nm"]
y = df["Power / dBm"]

fig, ax = plt.subplots()
plt.style.use("ITWM")
ax.plot(x, y)
ax.set_xlabel("Wavelength / nm", fontsize=10.5)
ax.set_ylabel("Power / dBm", fontsize=10.5)
ax.set_xlim(1532.5,1537.50)

ax.annotate('Center wavelength\n1534.8 nm \n\n3dB-bandwidth\n0.04 nm', xy=(260, 142), xycoords='axes points',
            size=9.5, ha='right', va='top',
            bbox=dict(boxstyle='round', fc='w'))

plt.savefig('../01_Plots/cwlaser_spectrum.png')
plt.show()
```
