### Parsing methods

```python
def StringIntoList(df, column):
    list_1 = []
    list_2 = []
    list_3 = []
    list_4 = []
    for n, i in enumerate(df[column]):
        list_1.append(df[column][n].replace('[', ''))
    for n, i in enumerate(list_1):
        list_2.append(list_1[n].replace(']', ''))
    for n, i in enumerate(list_2):
        list_3.append(list_2[n].split(','))
    for n, i in enumerate(list_3):
        list_h = []
        for a, b in enumerate(i):
            list_h.append(float(b))
        list_4.append(list_h)
    return list_4 

def StringIntoFloat(df,column):
    list_str = []
    list_float = []
    for n, i in enumerate(LD['current']):
        list_str.append(LD['current'][n].replace(',', '.'))
    for n, i in enumerate(ld_str):
        list_float.append(float(ld_str[n]))
    return ld_float

def StringIntoTmp(df,column):
    time_tmp_LD = []
    for n, i in enumerate(df[column]):
        time_tmp_LD.append(datetime.datetime.strptime(i,'%Y-%m-%d %H:%M:%S.%f'))
    return time_tmp_LD
    
def parseDataframe(df):
    # Drop first column
    df = df.drop(['Unnamed: 0'], axis=1)

    # Parse str into tmp
    df['time'] = StringIntoTmp(df,'time')

    # Parse str into lists
    df['x']= StringIntoList(df, 'x')
    df['y']= StringIntoList(df, 'y')
    df['x_max_NaN']= StringIntoList(df, 'x_max_NaN')
    df['y_max_NaN']= StringIntoList(df, 'y_max_NaN')
    df['x_max']= StringIntoList(df, 'x_max')
    df['y_max']= StringIntoList(df, 'y_max')
    df['delta_max']= StringIntoList(df, 'delta_max')

    # Set threshold for maxima detection
    max_thresh_1 = []
    for i in df['y_max_NaN']:
        max_thresh_2 = []
        for a,b in enumerate(i):
            if b < -56:
                max_thresh_2.append(np.nan)
            else:
                max_thresh_2.append(b)
        max_thresh_1.append(max_thresh_2)
    df['y_max_NaN'] = max_thresh_1

    return df

def addfreq(df):
    cenfreq_list = []
    kelminusfreq_list = []
    kelplusfreq_list = []
    c = scipy.constants.c

    for a,b in enumerate(df.cenwav):
        cenfreq_list.append(1e9*c/(b))
        kelminusfreq_list.append((1e9*c/(b))-(1e9*c/(b+df['kelly_-1'][a])))
        kelplusfreq_list.append((1e9*c/(b))-(1e9*c/(b+df['kelly_+1'][a])))
        #kelminusfreq_list.append((1e9*c/(b))-(1e9*c/(b+OSA_trace_final_1['kelly_-1'][a])))
        #kelplusfreq_list.append((1e9*c/(b))-(1e9*c/(b+OSA_trace_final_1['kelly_+1'][a])))

    df['cenfreq']=cenfreq_list
    df['kelminusfreq']=kelminusfreq_list
    df['kelplusfreq']=kelplusfreq_list

    return df
```

### Classes

```python
class Laser:
    def __init__(self, date, number, freq, netDispersion):
        # Path definitions
        folder = '../01_Data/DispersionMeasurement'
        
        self.date = date
        self.number = number
        self.freq = freq
        self.netDispersion = netDispersion
        self.code = date+'_'+number+'_'+freq
        self.OSATracePath = folder+'/'+self.code+'/'+self.code+'_'+'OSA_trace_final.txt'
        self.powerUpPath = folder+'/'+self.code+'/'+self.code+'_'+'power_up.tsv'
        
        # Reading & Parsing
        self.OSATrace = addfreq(parseDataframe(pd.read_csv(self.OSATracePath, sep="\t")))
        self.PowerUp = pd.read_csv(self.powerUpPath, sep="\t")
```

### 3dB Bandwidth

#### Fit arbitrary function

```python
from scipy.interpolate import UnivariateSpline
from scipy.optimize import leastsq

def threedB(current, plot=True):
    #fitfunc  = lambda p, x: p[0]*np.exp(-0.5*((x-p[1])/p[2])**2)+p[3]
    fitfunc  = lambda p, x: p[0]/(np.cosh((x-p[1])/p[2])**2)+p[3] ## just a function example --> MIND: Soliton has a sech shape in time domain, not wavelength domain!
    errfunc  = lambda p, x, y: (y - fitfunc(p, x))

    #fit
    current = current

    for a,b in enumerate(laser4.OSATrace['current']):
            c = current - b
            if abs(c) < 5.0:
                x = np.asarray(laser4.OSATrace['x'][a], dtype=None, order=None)
                y = np.asarray(laser4.OSATrace['y'][a], dtype=None, order=None)
                xdata = np.asarray(laser4.OSATrace['x'][a][int((1545-1516)/0.16):int((1574-1516)/0.16)], dtype=None, order=None)
                ydata = np.asarray(laser4.OSATrace['y'][a][int((1545-1516)/0.16):int((1574-1516)/0.16)], dtype=None, order=None)
                break

    init  = [1.0, 1559, 1, 1]

    out   = leastsq(errfunc, init, args=(xdata, ydata))
    c = out[0]
    
    if plot == True:
        # Plot fit
        fig, ax = plt.subplots()

        # Use our custom style
        plt.style.use('../01_Style/FigureTemplate_ITWM.mplstyle')

        plt.plot(x, y, label='measured data', color='#0000e4')
        plt.plot(xdata, fitfunc(c, xdata), label = 'fitted function', color='#d80000')

        plt.legend()          

        plt.xlim((1525,1595))
        plt.ylim((-70,-5))
        plt.ylabel('Intensity / dB', fontsize=10.5)
        plt.xlabel('Wavelength / nm', fontsize=10.5)

        #plt.savefig('../01_Plots/Fit.pdf', dpi=300, format='pdf')
        #plt.savefig('../01_Plots/Fit.png', dpi=300, format='png')

        plt.show()
        
    spline = UnivariateSpline(xdata, ydata-(np.max(ydata)-3), s=0)
    r1, r2 = spline.roots() # find the roots
    threedB = abs(r2-r1)
    print('3dB: ' + str(round(threedB, 1)) + ' nm')
    
    return threedB # 3dB Bandwidth /nm
```
